## Univariate Analysis vs Attrition – Categorical Variables

```python
# --------- Calculate attrition rate function ---------
def attrition_rate(df, col):
    return (
        df.groupby(col)["Attrition"]
        .apply(lambda x: (x == "Yes").mean())
        .reset_index(name="AttritionRate")
    )

# --------- Categories to analyze ---------
categories = [
    "Department",
    "JobRole",
    "BusinessTravel",
    "OverTime",
    "MaritalStatus",
    "Gender"
]

# --------- Plotting ---------
plt.figure(figsize=(10, 22))

for i, col in enumerate(categories, 1):
    plt.subplot(6, 1, i)
    
    rate_df = attrition_rate(df, col)
    
    sns.barplot(
        data=rate_df,
        x=col,
        y="AttritionRate"
    )
    
    plt.title(f"Attrition Rate (Yes) by {col}")
    plt.xlabel(col)
    plt.ylabel("Attrition Rate")
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

    1. Department vs Attrition Rate :
       Observation: Some departments show higher bars (higher attrition rates) than others.  
       Typical Pattern: In this dataset, Sales tends to have a slightly higher attrition rate compared to R&D, while HR                shows more volatility due to smaller employee counts.  
       Interpretation: Differences in department culture, nature of work, and management practices may influence employee              retention.

## Feature Engineering

```python
df["Attrition"] = df["Attrition"].map({"Yes": 1, "No": 0})
```

```python
df.head()
```

## Feature Selection

```python
numeric_features = [
    'Age', 'MonthlyIncome', 'TotalWorkingYears', 
    'YearsAtCompany', 'DistanceFromHome', 'JobLevel', 'StockOptionLevel'
]

categorical_features = [
    'OverTime', 'JobRole', 'MaritalStatus', 'BusinessTravel'
]

X = df[numeric_features + categorical_features]
y = df['Attrition']

# Split into Training and Testing sets (80/20 split)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)
```

```python
## One Hot Encoding
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer( 
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(drop='first'), categorical_features)
    ])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(
        n_estimators=100, 
        random_state=42, 
        class_weight='balanced'  # Addresses the imbalance in attrition data
    ))
])
```

```python
## Fit the model
pipeline.fit(X_train, y_train)
print("Pipeline trained successfully!")
```

```python
# Make predictions
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score
y_pred = pipeline.predict(X_test)
y_probs = pipeline.predict_proba(X_test)[:, 1]

# Calculate scores
accuracy = accuracy_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_probs)

print(f"--- Model Performance ---")
print(f"Accuracy Score: {accuracy:.2%}")
print(f"ROC-AUC Score:  {roc_auc:.2f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Example: Creating a dummy new employee record
new_employee = pd.DataFrame({
    'Age': [30],
    'MonthlyIncome': [2500],
    'TotalWorkingYears': [5],
    'YearsAtCompany': [4],
    'DistanceFromHome': [25],
    'JobLevel': [1],
    'NumCompaniesWorked': [4],
    'StockOptionLevel': [2],
    'OverTime': ['No'],
    'JobRole': ['Sales Representative'],
    'MaritalStatus': ['Married'],
    'BusinessTravel': ['Travel_Frequently']
})

# Predict using the pipeline
risk_score = pipeline.predict_proba(new_employee)[0][1]
print(f"Prediction: This employee has a {risk_score:.1%} probability of attrition.")
```
